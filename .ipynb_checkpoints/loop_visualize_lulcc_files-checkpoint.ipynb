```python
annual_current_veg = trendy.groupby(['year', 'current vegetation', 'disturbance type'])['cohort area '].sum().reset_index()
```

```python
len(annual_current_veg)
annual_current_veg
```

```python
annual_current_veg['current vegetation'].unique()
annual_current_veg.groupby(['current vegetation'])['disturbance type'].value_counts().reset_index(name='count')
annual_current_veg.groupby(['current vegetation', 'disturbance type'])['cohort area '].sum().reset_index()
```

```python
conditions = [
annual_current_veg['disturbance type'] ==2
, annual_current_veg['current vegetation'] == 50 
    ,annual_current_veg['current vegetation'].isin([51, 47])
, annual_current_veg['current vegetation'] == 48 
    ,annual_current_veg['current vegetation'].isin([4, 5, 6, 9, 10, 11, 16, 17, 18, 19, 20, 25, 33])
    ,annual_current_veg['current vegetation'].isin([7, 8, 12, 13, 14])
    ,annual_current_veg['current vegetation'].isin([15, 35])
    ,annual_current_veg['current vegetation'].isin([2, 3])
]

values = ['timber', 
          'crop', 
          'pasture', 
          'urban', 
          'forest', 
          'grass', 
          'shrub', 
          'tundra']

annual_current_veg['bens_groups'] = np.select(conditions, values, default = 'other')
```

```python
annual_current_veg['bens_groups'].value_counts()
annual_current_veg.loc[annual_current_veg['bens_groups'] == 'other', 'current vegetation'].value_counts()
```

```python
annual_current_veg
```

```python
plot_data = annual_current_veg.groupby(['bens_groups', 'year'])['cohort area '].sum().reset_index()
plot_data['percent_of_land'] = plot_data.groupby('year')['cohort area '].apply(lambda x: x / x.sum())
plot_data.loc[plot_data['year'] == 1700]
```

```python
# plot_data
```

```python
ax = sns.lineplot(data = plot_data, x='year', y = 'percent_of_land', hue = 'bens_groups', lw=3)
# sns.relplot(data=plot_data, x="year", y="passengers", hue="month", kind="line")

fmt = '%' # set the format as a percent
yticks = mtick.PercentFormatter(xmax=1.0, decimals=0, symbol=fmt) # create the formatter object
ax.yaxis.set_major_formatter(yticks) # set the formatter for the y-axis
plt.title('China - % of Land Cover by Year', fontdict={'fontsize': 14})
plt.ylabel('')
plt.xlabel('Year' , fontdict={'fontsize': 12})

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), fontsize=12)
plt.show()
```
